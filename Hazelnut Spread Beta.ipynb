## Sólidos de Revolución

Un sólido de revolución es un cuerpo obtenido mediante la rotación o revolución de un área plana alrededor de un eje del plano, normalmente llamado eje de revolución. A través de dicho proceso, podemos obtener formas tridimensionales, por ejemplo, al girar un triángulo rectángulo alrededor del eje, obtenemos un cono. Hay distintas formas para obtener el volúmen de estas formas, de las cuales cubriremos dos.

### Método de los Discos

Al igual que, al girar un triángulo rectángulo alrededor del plano, obtenemos un cono, al girar un rectángulo, podemos obtener un cilindro, o disco. Formando este disco, y calculando el volumen del mismo, podemos usarlo después para calcular el volumen de un sólido de revolución general. Para ello, primero debemos conocer la fórmula para calcular el volumen de un cilindro normal $v=h\,\pi\,r^{2}$. Debemos tomar una función $f\left(x\right)$ continua, y considerar una región delimitada para $y : \left[c,f\left(x\right)\right]$, y para $x : \left[a,b\right]$, girando, en este caso, alrededor de $c=0$.

Esta región será la misma que va a girar alrededor del eje de revolución, en este caso, el eje $x$, también representará nuestro radio $r$, en la fórmula del volumen del cilindro. Lo siguiente es crear los distintos discos que irán dentro del sólido de revolución, cuya suma se aproximará al volumen del mismo. Debemos generar rectángulos dentro del sólido de revolución, con un ancho igual a $x$, al girar estos rectángulos junto a nuestro sólido de revolución, obtendremos $n$ discos de radio $f\left(x\right)$ y $x$ de ancho, en donde $n$ es el número de particiones.

Para obtener la suma de los volúmenes de los discos ahora podemos usar una integral definida, ya que, cuando el número de discos tiende a infinito, más nos aproximamos al volumen real de nuestro sólido de revolución, asumiendo que nuestra $h$, en la fórmula del volumen del cilindro es igual al ancho de los cilindros. La integral definida resultante es la siguiente:

$$
v=\pi\int_{a}^{b} \left[F\left(x\right)\right]^{2}\,dx
$$

#### Ejemplos

Para $f\left(x\right)=x$, $y : \left[0,f\left(x\right)\right]$, y $x : \left[0,2\right]$

$$
\begin{split}
v&=\pi\int_{a}^{b} \left[F\left(x\right)\right]^{2}\,dx\\
&=\pi\int_{0}^{2} \left(x\right)^{2}\,dx\\
&=\pi\int_{0}^{2} \left(x^{2}\right)\,dx\\
&=\pi\,\left(\frac{1}{3}x^{3}\right)\big|_{0}^{2}\\
\therefore v&= \frac{8\pi}{3}u^{3}
\end{split}
$$

Para $f\left(x\right)=\sqrt{x}$, $y : \left[0,f\left(x\right)\right]$, y $x : \left[0,4\right]$

$$
\begin{split}
v&=\pi\int_{a}^{b} \left[F\left(x\right)\right]^{2}\,dx\\
&=\pi\int_{0}^{4} \left(\sqrt{x}\right)^{2}\,dx\\
&=\pi\int_{0}^{4} \left(x\right)\,dx\\
&=\pi\,\left(\frac{1}{2}x^{2}\right)\big|_{0}^{4}\\
\therefore v&= 8\pi u^{3}
\end{split}
$$

### Método de las Arandelas

Usando el método de los discos, encontramos siempre discos, sin ningún tipo de espacio vacío en el interior del cilindro, sin embargo, eso solo es así cuando el eje de revolución está tocando alguno de los límites de la función, el problema es que hay casos en el que queremos revolucionar por encima de ese límite, para eso usaremos el método de las arandelas. 

En este caso, en vez de obtener pequeños cilindros o discos, obtendremos arandelas, que son los mismo que discos con un cilindro atravesado, es decir, tienen un hoyo en el medio, y nuestro sólido de revolución será una esfera, semiesfera u otro tipo de forma, que ha sido atravesada por un cilindro.

Ahora en lugar de saber la fórmula del volumen del cilindro sin más, debemos tomar el volumen del cilindro sin tomar en cuenta el hueco, y restarle el volumen del cilindro que habría dejado ese hueco. Por lo que se restan los dos volúmenes.

$$
v = h\,\pi\,\left(R^{2}-r^{2}\right)
$$

siendo $R$ el radio del cilindro total y $r$ el radio del hueco.
Tomando nuestra función $f\left(x\right)$, delimitada para $y : \left[c,f\left(x\right)\right]$, y para $x : \left[a,b\right]$

La integral definida resultante es análoga a la anterior:

$$
v=\pi\int_{a}^{b} \left(R^{2}-r^{2}\right)\,dx
$$

#### Ejemplo

Para $f\left(x\right)=x^{2}$, $y : \left[2,f\left(x\right)\right]$, y $x : \left[-1,1\right]$, en donde $r$ es la Distancia entre el límite de la región especificada y el eje de revolución, y $R$ es la distancia entre el eje de revolución, y el eje $x$ menos el valor de $f\left(x\right)$ en un punto específico, para así obtener el radio de cualquier punto en la función.

$$
R = 2-x^{2} \land r=1
$$

$$
\begin{split}
\therefore v&=\pi\int_{a}^{b} \left(R^{2}-r^{2}\right)\,dx \\
&= \pi\int_{-1}^{1} \left[\left(2-x^{2}\right)^{2}-\left(1\right)^{2}\right]\,dx \\
v&= \pi\int_{-1}^{1} \left(3+x^{4}-4x^{2}\right)\,dx
\end{split}
$$

Ya que nuestra función es simétrica, y estamos tomando del punto menos uno al punto uno, podemos simplemente tomar el área de cero a uno y multiplicar la misma por dos

$$
\begin{split}
v&=\pi\int_{-1}^{1} \left(3+x^{4}-4x^{2}\right)\,dx\\
&= 2\pi\int_{0}^{1} \left(3+x^{4}-4x^{2}\right)\,dx\\
&= 2\pi(3x+\frac{1}{5}x^{5}-\frac{4}{3}x^{3})\big|_{0}^{1}\\
&= \frac{56\pi}{15}u^{3}
\end{split}
$$

Ahora, debido a que es muy tardado e ineficiente tratar de encontrar funciones que representen exactamente los bordes de la figura, vamos a utilizar el concepto de aproximaciones lineales

## Aproximaciones Lineales

Éstas se basan alrededor de las transformaciones encontradas en el teorema de Taylor, el cual se puede utilizar para encontrar un polinomio que sea la aproximación de cualquier otra función, lo cual resulta muy útil, ya que los polinomios son las ecuaciones más fáciles de manejar en cálculo. 

Aunque el teorema de Taylor produce una serie que se puede extender hasta tener un polinomio de grado $n$, nosotros nada más utilizaremos la parte lineal, pues nuestra aplicación no requiere de un grado tan alto de precisión, y al no tener límites en cuanto al número de divisiones que podemos hacer, cualquier otro grado sería tiempo perdido que podría ser mejor aprovechado haciendo otra cosa.<br>

$$
l\left(f\right) = f\left(a\right) + f'\left(a\right)\left(x-a\right) + O\left(x^{2}\right)
$$

## Precisión

Ahora, para estimar la precisión de nuestro modelo, vamos a comparar la capacidad de un cilindro de las dimensiones máximas y mínimas, y la capacidad del conenedor, determinada al llenarlo de agua, la cual sabemos que fue de $\approx1.125\,L$, y suponemos tiene una densidad de $\rho_{a}=1\,\tfrac{g}{ml}$. Sabemos que nuestro modelo toma en cuenta la tapa, por lo que el contenido sería un poco más alto.

Ésto lo haremos con la siguiente capacidad para los cilindros<br>

$$
C_{c}=h\,\pi\,r^{*}\ para \ r=r_{max}\ or\ r=r_{min}
$$

Y con la siguiente capacidad para el modelo<br>

$$
C_{m} = \pi\,\int_{h_{min}}^{h_{max}}\left[f\left(x\right)\right]^{2}\,dx
$$

En éste caso, para probar la precisión se integrará numéricamente.

**Redondeado a 4 decimales**

```python
integral=0 #variable for the sum, represents the integral
partitionSize = 100000 #Partition size for integrating, preferably matches the step used in x.linspaces(start,stop,step)
hmax = 17 #height of the bottle
hmin = 0 #base of the bottle
partition = (hmax-hmin)/partitionSize #defines the partition to be used

for i in vf(x):
    '''
    equivalent to the left riemann sum of the function, i guess, as i is equal to the function evaluated
    from 0 to 17 with the chosen step
    '''
    integral += (i**2)*partition

modelCapacity = ((integral)*np.pi) #Gives the total capacity of the container, in g; nested only so it is understood better

maxCylinderCapacity = ((np.pi*np.amax(vf(x))**2)*hmax) #Capacity of the largest cylinder, nested so it is better understood

minCylinderCapacity = ((np.pi*np.amin(vf(x))**2)*hmax)

print("The capacity of the biggest cylinder, with a radius of " + str(round(np.amax(vf(x)),4)) + "cm is " + str(round(maxCylinderCapacity,4)) +"ml")
print("The capacity of the smallest cylinder, with a radius of " + str(round(np.amin(vf(x)),4)) + "cm is " + str(round(minCylinderCapacity,4)) +"ml")
print("The capacity of our model is of " + str(round(modelCapacity,4)) + "ml")
```

La capacidad de nuestro modelo está entre la de los cilindros y solamente un poco mayor a la de 1.125L que medimos, por lo que creemos que nuestro modelo es lo suficientemente preciso para poder continuar con el proyecto.

## Volumen

El volumen de nuestra función se puede determinar a través de la fórmula que ya hemos usado para los sólidos de revolución.

**Todas las integrales y evaluaciones fueron generadas utilizando el modulo Sympy para python, en la parte del código viene cómo replicarlas, pero ocupaban mucho espacio, entonces fueron computadas por aparte**

$$
v=\pi\int_{a}^{b} \left[F\left(x\right)\right]^{2}\,dx
$$

Primero, tenemos que definir las integrales a usar para cada función de nuestra función original
$$
\begin{split}
&\int\limits_{0}^{0} 0\, dx \\
