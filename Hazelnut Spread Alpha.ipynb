```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import mpl_toolkits.mplot3d.axes3d as axes3d
import sympy as sp
#matplotlib.rcParams['figure.figsize'] = [10, 10]
plt.rcParams['figure.dpi'] = 150
```

```python
x=np.linspace(0,17,100000)
img = mpimg.imread('sidenut.png')

def f(x):
    if 0<=x<0.44:
        return 4.0307*x+1
    elif 0.44<=x<0.8:
        return 1.1275*x+2.2774
    elif 0.8<=x<1.12:
        return 0.6264*x+2.6783
    elif 1.12<=x<1.6:
        return 1.6045*x+1.5829
    elif 1.6<=x<1.95:
        return 1.11203*x + 2.37076
    elif 1.95<=x<2.52:
        return 0.55421*x + 3.4585
    elif 2.52<=x<3:
        return 0.04246*x + 4.74811
    elif 3<=x<3.2:
        return -0.37745*x + 6.00784
    elif 3.2<=x<11.88:
        return 0.02057*x + 4.73419
    elif 11.88<=x<12.11:
        return 0.44992*x - 0.36651
    elif 12.11<=x<12.2634:
        return -0.21464*x + 7.68128
    elif 12.2634<=x<14.236:
        return -0.46361*x + 10.73447
    elif 14.236<=x<14.4845:
        return -0.29342*x + 8.31168
    elif 14.4845<=x<14.7:
        return -0.01277*x + 4.24658
    elif 14.7<=x<14.89:
        return 1.52057*x - 18.29346
    elif 14.89<=x<16.83:
        return 0.00851*x + 4.22108
    elif 16.83<=x<=17:
        return -0.67978*x + 15.80506
    else:
        return 0.0
    
def cm_to_inch(value):
    return value/2.54
    
vf= np.vectorize(f)
#print(vf(x))
fig, ax= plt.subplots()
ax.plot(x,vf(x), color = "#30ff34", linewidth = "3")
ax.imshow(img, extent=[0, 17.2, 0, 5.2])
ax.set_title("Slido")
ax.set_ylim(0,6)
ax.set_xlim(0,17)
ax.set_xlabel("Alto (cm)")
ax.set_ylabel("Ancho (cm)")
plt.show
```
